# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
from scipy.stats import norm
import matplotlib.pyplot as plt
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
# Parmetros da distribuio:
mu = 15  # Mdia
sigma = 0.1  # Desvio padro
x = 15.2  # Limite superior

# Probabilidade de P(X > 15.2):
prob = 1 - norm.cdf(x, loc=mu, scale=sigma)

print(f"A probabilidade de que um pacote tenha mais que 15,2 kg : {prob:.4f}")
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Parmetros da distribuio:
mu = 15  # Mdia
sigma = 0.1  # Desvio padro
p = 0.95  # Percentil de ordem 95%

# Clculo do percentil:
percentile_95 = norm.ppf(p, loc=mu, scale=sigma)

print(f"O percentil de ordem 95% : {percentile_95:.4f} kg")
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
Y=np.linspace(-6, 6, 200)

# Cdigo:

from scipy.stats import norm, t

# Array de valores de Y:
Y = np.linspace(-6, 6, 200)

# Distribuio Normal Padro (mdia=0, desvio padro=1):
prob_normal = norm.cdf(Y, loc=0, scale=1)

# Distribuio t-Student com 5 graus de liberdade:
prob_t_student = t.cdf(Y, df=5)

# Exibindo os resultados para os primeiros 5 valores de Y:
for i in range(5):
    print(f"Y = {Y[i]:.2f}: P(Normal <= Y) = {prob_normal[i]:.4f}, P(t-Student <= Y) = {prob_t_student[i]:.4f}")
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
# Array de valores de Y:
Y = np.linspace(-6, 6, 200)

# Calculando as probabilidades acumuladas:
prob_normal = norm.cdf(Y, loc=0, scale=1)
prob_t_student = t.cdf(Y, df=5)

# Criando o grfico:
plt.figure(figsize=(10, 6))

# Plotando as distribuies:
plt.plot(Y, prob_normal, label="Normal Padro (N(0,1))", color="blue", linewidth=2)
plt.plot(Y, prob_t_student, label="t-Student (df=5)", color="orange", linestyle="--", linewidth=2)

# Adicionando ttulo e legendas:
plt.title("Comparao das Probabilidades Acumuladas", fontsize=16)
plt.xlabel("Valores de Y", fontsize=14)
plt.ylabel("P(X  Y)", fontsize=14)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)

# Exibindo o grfico:
plt.show()
```

5. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)

# Cdigo:

# Array de probabilidades acumuladas:
P = np.array([0.01, 0.05, 0.1, 0.5, 0.9, 0.95, 0.99])

# Calculando os percentis:
percentis_normal = norm.ppf(P)  # Percentis para Normal Padro
percentis_t_student = t.ppf(P, df=5)  # Percentis para t-Student (df=5)

# Exibindo os resultados:
for i, p in enumerate(P):
    print(f"P = {p:.2f}: Percentil Normal = {percentis_normal[i]:.4f}, Percentil t-Student = {percentis_t_student[i]:.4f}")
```

6. Faa um grfico com os dados do item anterior.

```python
# Array de probabilidades acumuladas:
P = np.array([0.01, 0.05, 0.1, 0.5, 0.9, 0.95, 0.99])

# Calculando os percentis:
percentis_normal = norm.ppf(P)  # Percentis para Normal Padro
percentis_t_student = t.ppf(P, df=5)  # Percentis para t-Student com 5 graus de liberdade

# Criando o grfico:
plt.figure(figsize=(10, 6))

# Plotando os percentis:
plt.plot(P, percentis_normal, label="Normal Padro (N(0,1))", marker="o", color="blue", linewidth=2)
plt.plot(P, percentis_t_student, label="t-Student (df=5)", marker="s", color="orange", linestyle="--", linewidth=2)

# Adicionando ttulos e legendas:
plt.title("Comparao dos Percentis entre Normal Padro e t-Student", fontsize=16)
plt.xlabel("Probabilidade Acumulada (P)", fontsize=14)
plt.ylabel("Percentis", fontsize=14)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)

# Exibindo o grfico:
plt.show()
```
